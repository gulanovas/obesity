```python
import pandas as pd
import matplotlib.pyplot as plt # matplot gives a lot of usefull function to visualize data
import numpy as np
import seaborn as sns # seaborn using to visualize data
import warnings
warnings.filterwarnings('ignore')
```

```python
# reading csv data by using pandas library
file_path = "../data/ObesityDataSet_raw_and_data_sinthetic.csv"
df = pd.read_csv(file_path)
df
```

```python
df.shape # check dataframe shape what gives values and columns lengh
```

```python
df[df.isnull().any(axis=1)] # checking if there no null data
```

```python
df.rename(columns = {'NObeyesdad':'Obesity'}, inplace = True) # change column name to be able easier to work
```

```python
df.dtypes # there is a lot of data which is float and have numbers after comma what I found not really usefull
df['Age'] = df['Age'].apply(np.int64) # it might be that number after comma present months and days what is not really usefull
df['FCVC'] = df['FCVC'].apply(np.int64) # easier to read
df['NCP'] = df['NCP'].apply(np.int64) # easier to read
df['FAF'] = df['FAF'].apply(np.int64) # easier to read
df['TUE'] = df['TUE'].apply(np.int64) # easier to read
df
```

```python
# A KDE plot giving more information about obesity weight range accoring to dataset. 
# It's quite hard to read but very usefull
# As possible to see there is a Insucfficient_Weight which are in the middle of somewhere
# For this reason next goal to make it in the beginning.
# Will use this to order types order=['Insufficient_Weight', 'Normal_Weight', 
# 'Overweight_Level_I', 'Overweight_Level_II', 
# 'Obesity_Type_I', 'Obesity_Type_II', 'Obesity_Type_III']

plt.figure(figsize=(14,6))

sns.boxplot(data=df, x='Obesity', y='Weight', order=['Insufficient_Weight', 'Normal_Weight', 'Overweight_Level_I',
                                                     'Overweight_Level_II', 'Obesity_Type_I', 
                                                     'Obesity_Type_II', 'Obesity_Type_III'])
```

```python
# pandas used to make weight and obesity columns values put to list
# matplotlib used to make chart size, title, labels and present legend of chart
# seaborn used to show chart averange(median) for weight according to obesity type
weight = df['Weight'].values.tolist()
obesity = df['Obesity'].values.tolist()

plt.figure(figsize=(14,6))


plt.title("Obesity type depends on weight")

# to avoid warning message instead of x,y changed to x=x, y=y 
# reference:
# (https://stackoverflow.com/questions/64130332/seaborn-futurewarning-pass-the-following-variables-as-keyword-args-x-y)
sns.lineplot(x=obesity, y=weight) 
legend_labels = ["Obesity type"]
plt.legend(legend_labels, loc=8)
plt.show()
```

```python
# Seperating eating habits and physical habits
eating_habits = df[['FAVC', 'FCVC', 'NCP', 'CAEC', 'CH2O', 'CALC']]
physical_habits = df[['SCC', 'FAF', 'TUE', 'MTRANS', 'SMOKE']]
habits = pd.concat([physical_habits, eating_habits.reindex(physical_habits.index)], axis=1)
habits
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() # Using label encoder to convert strings values to numerical values
dfle = df
dfle.Obesity = le.fit_transform(dfle.Obesity)
dfle.FAVC = le.fit_transform(dfle.FAVC)
dfle.CAEC = le.fit_transform(dfle.CAEC)
dfle.SMOKE = le.fit_transform(dfle.SMOKE)
dfle.SCC = le.fit_transform(dfle.SCC)
dfle.CALC = le.fit_transform(dfle.CALC)
dfle.MTRANS = le.fit_transform(dfle.MTRANS)
dfle.family_history_with_overweight = le.fit_transform(dfle.family_history_with_overweight)
dfle
```

```python
# Creating x values
X = dfle[['FAVC', 'FCVC', 'NCP', 'CAEC', 'CH2O', 'CALC', 'SCC', 'FAF', 'TUE', 'MTRANS', 'SMOKE']].values
X
```

```python
# Creating y values
y = dfle.Obesity
y
```

```python
# Before using OneHotEncoder need to convert string data to numerical by using Label Encoder
# One hot encoder find the unique values per feature and transform the data to a binary
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ohe = ColumnTransformer([("encoder", OneHotEncoder(), [1])], remainder = 'passthrough')
X = ohe.fit_transform(X)
X
```

```python
# Training models
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression 
# Additional libraries
from sklearn.model_selection import train_test_split # Used to test data
from sklearn import preprocessing # Used to preprocess data
from sklearn.pipeline import make_pipeline # Used to contruct pipelines
```

```python
# LogisticRegression model
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=1, stratify=y)
scaler = preprocessing.StandardScaler().fit(X_train)
model = make_pipeline(scaler, LogisticRegression())
model.fit(X_train, y_train.values.ravel())
# make as whole percentage
lr = round(model.score(X_test, y_test) * 100)
```

```python
# DecisionTreeClassifier model
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=1, stratify=y)
scaler = preprocessing.StandardScaler().fit(X_train)
model = make_pipeline(scaler, DecisionTreeClassifier())
model.fit(X_train, y_train.values.ravel())
# make as whole percentage
dtc = round(model.score(X_test, y_test) * 100)
```

```python
# GaussianNB model
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=1, stratify=y)
scaler = preprocessing.StandardScaler().fit(X_train)
model = make_pipeline(scaler, GaussianNB())
model.fit(X_train, y_train.values.ravel())
# make as whole percentage
gnb = round(model.score(X_test, y_test) * 100)
```

```python
# SVC model
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=1, stratify=y)
scaler = preprocessing.StandardScaler().fit(X_train)
model = make_pipeline(scaler, SVC())
model.fit(X_train, y_train.values.ravel())
# make as whole percentage
svc =  round(model.score(X_test, y_test) * 100)
```

```python
# That's shows the percentage of every model accuracy
predictions = pd.DataFrame({'predictions':['LogisticRegression', 'DecisionTreeClassifier', 'GaussianNB', 'SVC'],
                     'percentage':[lr, dtc, gnb, svc]},
                      index= ['LogisticRegression', 'DecisionTreeClassifier', 'GaussianNB', 'SVC'])
predictions
```

```python
# this pie-chart helps to see which graph were most acurracy
predictions.plot(kind='pie', y='percentage', autopct='%1.0f%%', figsize=(7,7))
```
